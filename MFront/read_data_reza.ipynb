```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# List of CSV files
csv_files = [f for f in os.listdir('Results_Reza/') if f.endswith('.csv')]

# List to store DataFrames
df_list = []

# Loop through the list of files
for file in csv_files:
    df = pd.read_csv('Results_Reza/'+file)
    df_list.append(df)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df['AxialStrainHistory'].index,df['AxialStrainHistory'])
ax.set_xlabel('Pseudo-time')
ax.set_ylabel('$\\epsilon_\\mathrm{ax}$');
```

```python
fig, ax = plt.subplots(figsize=(12,12),nrows=3)
for df in df_list:
    ax[0].plot(df['AxialStrainHistory'], df['HydrostaticPressure']/1e3,alpha=0.1)
    ax[0].set_ylabel("$p'$ / kPa")
    ax[0].set_xlabel("$\\epsilon_\\mathrm{ax}$")
    #
    ax[1].plot(df['AxialStrainHistory'], df['DeviatoricStress']/1e3,alpha=0.1)
    ax[1].set_ylabel("$q$ / kPa")
    ax[1].set_xlabel("$\\epsilon_\\mathrm{ax}$")
    #
    ax[2].plot(df['AxialStrainHistory'], df['VolumeStrain'],alpha=0.1)
    ax[2].set_ylabel("$\\epsilon_\\mathrm{vol}$")
    ax[2].set_xlabel("$\\epsilon_\\mathrm{ax}$")

fig.tight_layout()
fig.savefig('sample_results.pdf')
```
